### Задание 1

```python
import numpy as np
import pandas as pd
```

```python
authors = pd.DataFrame({
    'author_id': [1, 2, 3],
    'author_name': ['Turgenev', 'Chekhov', 'Ostrovskiy'],
})

authors
```

```python
books = pd.DataFrame({
    'author_id': [1, 1, 1, 2, 2, 3, 3],
    'book_title': ['Fathers and Sons', 'Rudin', 'The Noble Nest', 'Thick and thin', 'Lady with a dog', 
                   'The Thunderstorm', 'Talents and fans'],
    'price': [450, 300, 350, 500, 450, 370, 290]
})

books
```

### Задание 2

```python
authors_price = pd.merge(authors, books, on='author_id', how='outer')

authors_price
```

### Задание 3

```python
top5 = authors_price.nlargest(5, "price")

top5
```

### Задание 4

```python
authors_stat = pd.DataFrame({
    'author_name': authors_price["author_name"],
    'min_price': authors_price["price"],
    'max_price': authors_price["price"],
    'mean_price': authors_price["price"],
})

authors_stat.groupby("author_name").agg({"min_price": "min", "max_price": "max", "mean_price": "mean"})
```

### Задание 5

```python
authors_price["cover"] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']

authors_price
```
